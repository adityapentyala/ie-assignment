```python
import seaborn as sb
import pandas as pd
```

```python
df=pd.read_excel('data/IE_fnl_fnl_data.xlsx')

df.head()
```

```python
df.drop(['year','country_id_o','country_id_d'],axis=1,inplace=True)
df.head()
```

```python
corr_matrix=df.corr(numeric_only=True)
corr_matrix=corr_matrix.round(3)
```

```python
dataplot=sb.heatmap(corr_matrix,cmap="YlGnBu", annot=True)
```
